```
def convert(DF,SortForm,Actual,Column):
  for i in range(len(DF[Column])):
    for j in range(len(SortForm)):
      if SortForm[j] == DF.loc[i,Column]:
        DF.loc[i,Column] = Actual[j]
  return DF
```

```
def catagoryCov(DF,Column,drop_first):
  dummis = pd.get_dummies(DF[Column],drop_first=drop_first)
  marge = pd.concat([DF,dummis],axis='columns')
  marge = marge.drop([Column],axis='columns')
  return marge
```

```
cols = ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color',
       'ring-type', 'spore-print-color', 'population', 'habitat']
```

```
DF = df
for col in cols:
  DF = catagoryCov(DF,col,False)
DF.columns
```

```
DF = catagoryCov(DF,'class')
DF
```

```
DF = DF.drop(['NaN'],axis='columns')
DF
```

```
X = DF.iloc[:,:-1]
y = DF.iloc[:,-1]
```

```
X
```

```
y
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.1, random_state=1)
```

```
from xgboost import  XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train.values,y_train.values)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test.values)
cm = confusion_matrix(y_test.values,y_pred)
print(cm)
accuracy_score(y_test.values,y_pred)
```
